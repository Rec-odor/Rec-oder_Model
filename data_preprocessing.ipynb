```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# 향수 데이터셋 불러오기
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/capstone/data/data.csv', encoding='utf-8')
df.head(1)
```

```python
# Top Note, Middle Note, Base Note를 모두 합친 Notes 칼럼 생성
df['Notes'] = df['Top Note'].str.cat(df['Middle Note'], sep=',')
df['Notes'] = df['Notes'].str.cat(df['Base Note'], sep=',')
df["Notes"] = df['Notes'].str.strip(",")
df.head(1)
```

```python
# Notes list로 변환
df['Notes_list'] = df.Notes.str.split(',')
df['Notes_list'] = df['Notes_list'].apply(lambda l: list(set([e for e in l if e != ""])))
df.head(1)
```

```python
# list 공백 제거
for i in range(len(df)):
    df.at[i, 'Notes_list'] = [i.strip() for i in df.at[i, 'Notes_list']]

df.head(1)
```

```python
# multi-hot encoding 진행
mlb = MultiLabelBinarizer()
arr = mlb.fit_transform(df['Notes_list'])
```

```python
# unique한 향 개수
len(mlb.classes_)
```

```python
# multi-hot df 데이터프레임에 추가
df = df.assign(multi_hot = arr.tolist())
df.head(1)
```

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/capstone/data/data_preprocessing.csv',index=False)
```
