```python
import gensim
from gensim.models import Word2Vec 
from gensim.models import KeyedVectors
import pandas as pd

from copy import deepcopy
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
word_vec = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True, unicode_errors='ignore')
```

```python
# subgroup 벡터값 저장
woody = word_vec['woody']
fresh = word_vec['fresh']
floral = word_vec['floral']
oriental = word_vec['oriental']
```

```python
top_dict, middle_dict, base_dict = {}, {}, {}

with open("top_dict_new", "r") as f:
    for line in f.readlines():
        if line:
            v = line.strip().split(", ")
            top_dict[v[0]] = v[1].lower().replace("*", "").split()


with open("middle_dict_new", "r") as f:
    for line in f.readlines():
        if line:
            v = line.strip().split(", ")
            try:
                middle_dict[v[0]] = v[1].lower().replace("*", "").split()
            except:
                print(v)
                raise 'o'

with open("base_dict_new", "r") as f:
    for line in f.readlines():
        if line:
            v = line.strip().split(", ")
            base_dict[v[0]] = v[1].lower().replace("*", "").split()
```

```python
top_vec = {}

for k, l in top_dict.items():
    vecs = None
    for v in l:
        try:
            if vecs is None:
                vecs = deepcopy(word_vec[v])
            else:
                vecs += deepcopy(word_vec[v])
        except:
            continue
    if vecs is not None:
        top_vec[k] = vecs
    else:
        top_vec[k] = np.zeros(300)
        print("벡터값 0!", k, l)

middle_vec = {}

for k, l in middle_dict.items():
    vecs = None
    for v in l:
        try:
            if vecs is None:
                vecs = deepcopy(word_vec[v])
            else:
                vecs += deepcopy(word_vec[v])
        except:
            continue
    if vecs is not None:
        middle_vec[k] = vecs
    else:
        middle_vec[k] = np.zeros(300)
        print("벡터값 0!", k, l)
        
base_vec = {}

for k, l in base_dict.items():
    vecs = None
    for v in l:
        try:
            if vecs is None:
                vecs = deepcopy(word_vec[v])
            else:
                vecs += deepcopy(word_vec[v])
        except:
            continue
    if vecs is not None:
        base_vec[k] = vecs
    else:
        base_vec[k] = np.zeros(300)
        print("벡터값 0!", k, l)
```

```python
top_subgroup = {}

for key, value in top_vec.items():
    sub = []
    woody_ = cosine_similarity(woody.reshape(1, -1), value.reshape(1, -1))
    fresh_ = cosine_similarity(fresh.reshape(1, -1), value.reshape(1, -1))
    floral_ = cosine_similarity(floral.reshape(1, -1), value.reshape(1, -1))
    oriental_ = cosine_similarity(oriental.reshape(1, -1), value.reshape(1, -1))
    sub.append(woody_)
    sub.append(fresh_)
    sub.append(floral_)
    sub.append(oriental_)
    num = sub.index(max(sub))
    if num==0:
        group = 'woody'
    elif num==1:
        group = 'fresh'
    elif num==2:
        group = 'floral'
    else:
        group = 'oriental'
    top_subgroup[key] = group
```

```python
middle_subgroup = {}

for key, value in middle_vec.items():
    sub = []
    woody_ = cosine_similarity(woody.reshape(1, -1), value.reshape(1, -1))
    fresh_ = cosine_similarity(fresh.reshape(1, -1), value.reshape(1, -1))
    floral_ = cosine_similarity(floral.reshape(1, -1), value.reshape(1, -1))
    oriental_ = cosine_similarity(oriental.reshape(1, -1), value.reshape(1, -1))
    sub.append(woody_)
    sub.append(fresh_)
    sub.append(floral_)
    sub.append(oriental_)
    num = sub.index(max(sub))
    if num==0:
        group = 'woody'
    elif num==1:
        group = 'fresh'
    elif num==2:
        group = 'floral'
    else:
        group = 'oriental'
    middle_subgroup[key] = group
```

```python
base_subgroup = {}

for key, value in base_vec.items():
    sub = []
    woody_ = cosine_similarity(woody.reshape(1, -1), value.reshape(1, -1))
    fresh_ = cosine_similarity(fresh.reshape(1, -1), value.reshape(1, -1))
    floral_ = cosine_similarity(floral.reshape(1, -1), value.reshape(1, -1))
    oriental_ = cosine_similarity(oriental.reshape(1, -1), value.reshape(1, -1))
    sub.append(woody_)
    sub.append(fresh_)
    sub.append(floral_)
    sub.append(oriental_)
    num = sub.index(max(sub))
    if num==0:
        group = 'woody'
    elif num==1:
        group = 'fresh'
    elif num==2:
        group = 'floral'
    else:
        group = 'oriental'
    base_subgroup[key] = group
```

```python
# list로 변환
df['Top Note'] = df["Top Note"].apply(lambda s: s.split(", ")).apply(lambda l: [i.strip() for i in l])
df['Middle Note'] = df["Middle Note"].apply(lambda s: s.split(", ")).apply(lambda l: [i.strip() for i in l])
df['Base Note'] = df["Base Note"].apply(lambda s: s.split(", ")).apply(lambda l: [i.strip() for i in l])
```

```python
# subgroup dictionay와 매핑
df['Top_Subgroup'] = df["Top Note"].apply(lambda l: list(map(lambda x: top_subgroup[x], l)))
df['Middle_Subgroup'] = df["Middle Note"].apply(lambda l: list(map(lambda x: middle_subgroup[x], l)))
df['Base_Subgroup'] = df["Base Note"].apply(lambda l: list(map(lambda x: base_subgroup[x], l)))
```

```python
df['Subgroup'] = df['Top_Subgroup'] + df['Middle_Subgroup'] + df['Base_Subgroup']
df.head(1)
```

```python
# 각 분류에 해당하는 개수 세기
woody_cnt = df['Subgroup'].apply(lambda l: [i.strip() for i in l]).apply(lambda x: x.count('woody'))
fresh_cnt = df['Subgroup'].apply(lambda l: [i.strip() for i in l]).apply(lambda x: x.count('fresh'))
floral_cnt = df['Subgroup'].apply(lambda l: [i.strip() for i in l]).apply(lambda x: x.count('floral'))
oriental_cnt = df['Subgroup'].apply(lambda l: [i.strip() for i in l]).apply(lambda x: x.count('oriental'))
subgroup_len = df['Subgroup'].apply(lambda l: [i.strip() for i in l]).apply(lambda x: len(x))
```

```python
# percentage 계산
woody_percent = round(woody_cnt/subgroup_len*100, 2)
fresh_percent = round(fresh_cnt/subgroup_len*100, 2)
floral_percent = round(floral_cnt/subgroup_len*100, 2)
oriental_percent = round(oriental_cnt/subgroup_len*100, 2)
```

```python
woody_percent = woody_percent.apply(lambda n: str(n)).apply(lambda s: s + '%').apply(lambda l: 'woody: ' + l)
fresh_percent = fresh_percent.apply(lambda n: str(n)).apply(lambda s: s + '%').apply(lambda l: 'fresh: ' + l)
floral_percent = floral_percent.apply(lambda n: str(n)).apply(lambda s: s + '%').apply(lambda l: 'floral: ' + l)
oriental_percent = oriental_percent.apply(lambda n: str(n)).apply(lambda s: s + '%').apply(lambda l: 'oriental: ' + l) 
```

```python
total_percent = woody_percent + ', ' + fresh_percent + ', ' + floral_percent + ', ' + oriental_percent
```

```python
df['percentage(%)'] = total_percent
df
```
